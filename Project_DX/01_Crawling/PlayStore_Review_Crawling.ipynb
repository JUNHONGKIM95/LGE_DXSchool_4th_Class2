# 1. 구글 플레이스토어

### 앱 리뷰 페이지 정보 출력

- 키즈노트 :        com.vaultmicro.kidsnote
- 하이클래스 :      com.iscreammedia.app.hiclass.android
- 열나요 :          com.appmd.fever
- 베이비타임 :      yducky.application.babytime
- 베이비빌리 :      kr.villagebaby.billyapplication
- 마미톡 :          com.mommytalkappv1

```python
appList = ['키즈노트', '하이클래스', '열나요', '베이비타임', '베이비빌리', '마미톡', '280days', '삐요로그', '맘스다이어리', '쑥쑥찰칵', '베베메모']
appIdList = ['com.vaultmicro.kidsnote',
             'com.iscreammedia.app.hiclass.android',
             'com.appmd.fever',
             'yducky.application.babytime',
             'kr.villagebaby.billyapplication',
             'com.mommytalkappv1',
             'com.amanefactory.totsukitoka',
             'jp.co.sakabou.piyolog',
             'com.moms.momsdiary',
             'com.jejememe.daybabyday',
             'com.liveyap.timehut.bbmemo'
             ]
```

```python
appListCnt = len(appList)
appListCnt
```

### Macro

```python
from google_play_scraper import app
from google_play_scraper import Sort, reviews
import pandas as pd
from tqdm import tqdm

for idx in tqdm(range(appListCnt)):
    
    # 1. 앱 정보 입력
    result = app(
        appIdList[idx], #'com.samsung.android.oneconnect'
        lang='ko', # defaults to 'en'
        country='kr' # defaults to 'us'
    )
    
    reviewCnts = result['reviews']
    print('전체 리뷰 수: ', reviewCnts)
    
    # 2. 앱 리뷰 데이터 추출
    result, continuation_token = reviews(
        appIdList[idx],
        lang='ko', # defaults to 'en'
        country='kr', # defaults to 'us'
        sort=Sort.NEWEST, # defaults to Sort.NEWEST
        count=reviewCnts, # defaults to 100
        filter_score_with=None # defaults to None(means all score)
    )
    
    # 3. 데이터프레임 전처리
    df = pd.DataFrame(result)
    nCnt = len(df)
    
    # 컬럼 추가
    df2 = df[['content', 'at']]
    df2.insert(0, '출처', '플레이스토어_%s'%(appList[idx]))
    df2.insert(1, '제목', '-')
    df2.insert(4, '링크', '-')

    # 컬럼명 변경 (content -> 내용, at -> 작성일)
    # 기존에 '제목' 컬럼이 이미 존재하므로, content는 '내용'으로 변경하는 것이 중복을 피할 수 있습니다.
    df2 = df2.rename(columns={'content': '내용', 'at': '작성일'})

    # '작성일' 컬럼을 datetime 형식으로 변환 후 YYYY-MM-DD 문자열로 변경
    df2['작성일'] = pd.to_datetime(df2['작성일']).dt.strftime('%Y-%m-%d')

    
    # 4. csv 저장
    df2.to_csv("./crawled/플레이스토어_%s(%s건).csv"%(appList[idx], nCnt), index=False, encoding='utf-8-sig')
```
