```python
import os
import pandas as pd
import pickle
from tqdm import tqdm
import ast
```

```python
# 1. 개별 데이터프레임을 담을 빈 리스트 생성
df_list = []

for i in range(5):
    # f-string을 이용해 파일 읽기
    df = pd.read_csv(f'./results/LDA_Actor{i}_action_cluster.csv')
    # 리스트에 추가
    df_list.append(df)

# 2. 리스트에 담긴 모든 데이터프레임을 수직(axis=0)으로 합치기
final_df = pd.concat(df_list, ignore_index=True)

```

```python
# 결과 확인
print(len(final_df))
final_df.head()
```

```python
# 1. 컬럼명 변경 (기존 'cluster' -> 'actor_cluster')
final_df = final_df.rename(columns={'cluster': 'actor_cluster'})

# 2. 결과 확인
print(len(final_df))
final_df.head()
```

```python
# 'actor_cluster'와 'action_cluster' 컬럼의 결측치 개수 확인
missing_values = final_df[['actor_cluster', 'action_cluster']].isnull().sum()

print(missing_values)
```

```python
# csv 저장
final_df.to_csv(f'./results/LDA_Actor_Action_cluster_integrated.csv', encoding='utf-8-sig')
```
