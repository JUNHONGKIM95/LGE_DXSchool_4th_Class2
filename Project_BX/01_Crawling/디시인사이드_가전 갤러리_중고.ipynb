```python
# 6. 코드 통합 (한번의 실행으로 전체 코드 작동)

# keyword = quote('중고')
keyword = quote('')
option = 999999 # page

cnts = 0

# f = open('디시인사이드_가전 갤러리_크롤링_중고.txt', 'w', encoding='utf-8')
f = open('디시인사이드_가전 갤러리_크롤링.txt', 'w', encoding='utf-8')

# driver를 통하여 페이지 요청하기
driver = wb.Chrome()
time.sleep(1)

for p in range(1, option+1):
    print('Page: ', p)
    
    try:
        url = f'https://gall.dcinside.com/board/lists/?id=appliance&page={p}&search_pos=-8396&s_type=search_subject_memo&s_keyword={keyword}'

        driver.get(url)
        time.sleep(1)
        
        # 4. 지식인 주소 수집하기 (href_list)
        contents = driver.find_elements(By.CSS_SELECTOR, 'tr.ub-content.us-post > td.gall_tit.ub-word > a')
        href_list = [c.get_attribute('href') for c in contents]
        title_list = [c.text for c in contents]

        print("게시글(url) 개수: ", len(href_list))
        print("첫 게시글 url: ", href_list[0])


        # 5. 블로그 접근하여 데이터 수집하기
        # for href in tqdm(href_list):
        for href in tqdm(href_list):
            cnts+= 1
            driver.get(href)
            time.sleep(0.5)
            
            # iframe
            # driver.switch_to.frame('mainFrame')
            
            # content_ = driver.find_element(By.CSS_SELECTOR, 'div.se-main-container')
            content_ = driver.find_element(By.CSS_SELECTOR, 'div.write_div')
            #print(content_.text)
            
            f.write(content_.text)
    except:
        print('Finished')
        break

f.close()
driver.quit()

print("전체 게시글 개수: ", cnts)
```

##### 7. 워드클라우드 시각화

```python
from kiwipiepy import Kiwi # 형태소 분석기
from collections import Counter # 단어가 나온 횟수를 셈
from wordcloud import WordCloud # 워드클라우드 생성 도구
import matplotlib.pyplot as plt # 시각화 도구
from PIL import Image
import numpy as np
```

```python
# f = open('디시인사이드_가전 갤러리_크롤링_중고.txt', 'r', encoding='utf-8')
f = open('디시인사이드_가전 갤러리_크롤링.txt', 'r', encoding='utf-8')
review = f.read()
f.close()
print(len(review))
#print(review)
```

```python
# 도구 객체생성
kiwi = Kiwi()

# 토큰화
token = kiwi.tokenize(review)
#token

# form: 토큰화 결과
# tag: 품사 태그
# nnp: 고유명사; nng: 일반명사
```

```python
# 일반명사만 추출 (NNG)
# 리스트명 = [실행문 반복문 조건]
# nng_token = [t.form for t in token if t.tag=='NNG']
nngp_token = [t.form for t in token if t.tag=='NNG' or t.tag=='NNP']

print(len(nngp_token))
#print(nngp_token)
```

```python
# 2글자 이상의 단어만 추출
nngp_token2 = [t for t in nngp_token if len(t)>=2]
nngp_token2
```

```python
# 명사의 개수를 세서 많이 나오는 단어를 활용하여 워드클라우드 생성하기
counter = Counter(nngp_token2)
counter
```

```python
# 상위 N개 단어만 추출
top_N = counter.most_common(50)
top_N
```

```python
# WordCloud(): 스타일(배경,글꼴), 최대단어수, 마스크이미지 등 옵션을 설정
# generate_from_frequencies(): 미리 정의된 단어의 빈도수를 이용해서 워드클라우드 이미지를 생성

wc = WordCloud(
    font_path='C:/Windows/Fonts/LG PC.ttf', # malgunbd.ttf
    background_color='white'
    # mask=mask
        
).generate_from_frequencies(dict(top_N))

plt.imshow(wc)
plt.axis('off')
# plt.show()

plt.savefig('디시인사이드_가전 갤러리_크롤링.png')
plt.close()
```
