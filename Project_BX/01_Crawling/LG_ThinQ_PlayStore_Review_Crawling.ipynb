### 앱 리뷰 페이지 정보 출력

```python
from google_play_scraper import app

result = app(
    'com.lgeha.nuts', #'com.samsung.android.oneconnect'
    lang='ko', # defaults to 'en'
    country='kr' # defaults to 'us'
)
result
```

```python
reviewCnts = result['reviews']
print('전체 리뷰 수: ', reviewCnts)
```

### 리뷰 데이터 추출

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.lgeha.nuts',
    lang='ko', # defaults to 'en'
    country='kr', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=reviewCnts, # defaults to 100
    filter_score_with=None # defaults to None(means all score)
)

# If you pass `continuation_token` as an argument to the reviews function at this point,
# it will crawl the items after N review items.
"""
result, _ = reviews(
    'com.lgeha.nuts',
    continuation_token=continuation_token # defaults to None(load from the beginning)
)
"""
```
